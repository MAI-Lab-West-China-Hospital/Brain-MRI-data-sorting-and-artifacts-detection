```python
import os
import torch
from torch.utils.data import DataLoader
import numpy as np
import matplotlib.pyplot as plt
from glob import glob
import cv2

from monai.visualize import CAM, GradCAM
from monai.transforms import *

from dataset import MRIDataset
from EfficientNet import MyEfficientNet
```

```python
# load model
device = torch.device("cpu")

model = MyEfficientNet("efficientnet-b0", spatial_dims=2, in_channels=3,
                        num_classes=13, pretrained=False, dropout_rate=0.2).to(device)
save_path = '../Results/EfficientNet/20230226_stack3_lr1e-05_epochs100_pretrainTrue'
model_path = os.path.join(save_path, 'Saved_models/best_model.pth')
model.load_state_dict(torch.load(model_path))
model.eval()

print('load model')
```

```python
print(model)
```

```python
# get module name
for name, _ in model.named_modules():
    print(name)
```

```python
cam = CAM(nn_module=model, target_layers='backbone._bn1', fc_layers='backbone._fc.0')
gradcam = GradCAM(nn_module=model, target_layers='backbone._bn1')
```

```python
### dataset ####
val_transforms = Compose(
        [
            ScaleIntensityd(keys='img'),
            EnsureTyped(keys='img')
        ]
    )
imgs = glob('../MRISeq_classification/CAM_imgs/*.gz')
data = []
for i in imgs:
    data.append({'img': i, 'label': 0})   # label is not importance
ds = MRIDataset(data=data, stack=3, transform=val_transforms)
loader = DataLoader(ds, batch_size=1, num_workers=1)
```

```python
# get cam
cams = []
names = []
images = []
for d in loader:
    input, labels = d["img"].to(device), d["label"].to(device)
    #print("test1")
    names.append(d['name'])
    #print("test2")
    cam_ =cam(input)
    #print("test3")
    cam_ = cam_.numpy().squeeze()
    #print("test4")
    image = d["img"].numpy().squeeze()
    #print("test5")
    images.append(image)
    #print("test6")
    cams.append(cam_)
    #print("test7")
```

```python
#show cam on image
cam_one = cams[1]
im = images[1]

weight = 0.6
plt.figure(figsize=(16,8))
for sub in range(3):
    plt.subplot(2, 3, sub+1)
    image_one = im[sub, ...]
    
    cam_ = cam_one[:, ::-1].T
    image_ = image_one[:, ::-1].T
    
    cam_image = cam_ * (1-weight)+ image_* weight
    cam_image = cam_image # / np.max(cam_image)
    
    plt.imshow(cam_image, cmap='jet')
    plt.axis('off')
plt.colorbar()
```
